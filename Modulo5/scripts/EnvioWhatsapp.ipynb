```python
import pandas as pd, datetime, pywhatkit
```

```python
df = pd.read_excel('ALUMNOS PYTHON-J1.xlsx')
print(df.head())
```

```python
columnas_seleccionadas = ['Nombre y apellidos','Número de celular','Nota final']
df = df[columnas_seleccionadas]
df = df.round({'Nota final':1})
print(df.head())
```

```python
#excluir filas con notas igual a cero
condicion=(df['Nota final']!=0)
df=df[condicion]
print(df.shape)
```

```python

for row in df.itertuples(index=False,name=None):
    #print(row[0],'-',row[1])
    #print(f'Hola {row[0]}, tu nota final es {round(row[2],1)}')
    
    try:
        pywhatkit.sendwhatmsg('+51'+str(row[1])
                              ,f'Hola {row[0]}, tu nota final es {round(row[2],1)}'
                              ,datetime.datetime.now().hour
                              ,datetime.datetime.now().minute+1
                              ,20,True,2
                              )
    except Exception as e:
        print(f'[{datetime.datetime.now()}] Error al enviar el mensaje: {e}')
        with open('logs_envíos.txt','a') as f:
            f.write(f'{datetime.datetime.now()} - FALLO - +51{row[1]}, Error: {e}')
print('Concluido')
```
