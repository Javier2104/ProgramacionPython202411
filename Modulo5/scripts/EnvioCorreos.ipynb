```python
import pyodbc, pandas as pd, smtplib
from email.message import EmailMessage
```

```python
df=pd.read_excel('ALUMNOS PYTHON-J1.xlsx')
print(df.head())
```

```python
df = df.drop(['Número de celular','Curso a matricularse'],axis=1)
df=df.round({'Nota final':1})
df=df.fillna(0)
print(df.head())
```

```python
df=df.rename(columns={
    df.columns[0]:'Nombres_apellidos',
    df.columns[1]:'Email',
    df.columns[2]:'Nota_final',
    df.columns[3]:'Nota1',
    df.columns[4]:'Nota2',
    df.columns[5]:'Nota3',
    df.columns[6]:'Nota4',
    df.columns[7]:'Nota5',
    df.columns[8]:'Nota6',
    df.columns[9]:'Nota7'
})

print(df.shape)
print(df.head())
```

```python
condicion = (df['Nota_final']!=0)
df=df[condicion]
print(df.shape)
```

```python
#Configuración de email
smtp_server='smtp.gmail.com'
smtp_port=587
email_sender='aguillenv.uni@gmail.com'
email_password='abcd abcd abcd abcd' #acá colocar la clave de aplicación que les aparece en la página https://myaccount.google.com/apppasswords
```

```python
def enviar_correo(Nombres_apellidos, Email, Nota_final,Nota1,Nota2,Nota3,Nota4,Nota5,Nota6,Nota7):
    msg=EmailMessage()
    msg['Subject']='correo de prueba'
    msg['From']=email_sender
    msg['To']=Email

    mensaje=f'''
    {Nombres_apellidos},

    Este es un mensaje de prueba.
    Nota final: {Nota_final}
    Nota1: {Nota1}
    Nota2: {Nota2}
    Nota3: {Nota3}
    Nota4: {Nota4}
    Nota5: {Nota5}
    Nota6: {Nota6}
    Nota7: {Nota7}
    
    Saludos,
    '''

    msg.set_content(mensaje)

    try:
        with smtplib.SMTP(smtp_server,smtp_port) as server:
            server.starttls()
            server.login(email_sender,email_password)
            server.send_message(msg)
        print(f'Correo enviado a {Email}')
    except Exception as e:
        print(f'Error al enviar correo a {Email}: {e}')
```

```python
for _,row in df.iterrows():
    enviar_correo(row['Nombres_apellidos'],row['Email'],row['Nota_final'],row['Nota1'],row['Nota2'],row['Nota3'],row['Nota4'],row['Nota5'],row['Nota6'],row['Nota7'])
```
